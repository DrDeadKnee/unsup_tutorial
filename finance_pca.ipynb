```python
allticks = pd.DataFrame()
def add_ticker(allticks, ticker, startdate="2018-01-01"):
    hist = yf.Ticker(ticker).history(start=startdate).reset_index(drop=False)
    slim = hist[["Date", "Open"]]
    slim["ticker"] = ticker
    allticks = allticks.append(slim, ignore_index=True)
    return allticks

allticks = add_ticker(allticks, "MSFT")
add_ticker(allticks, "GOOG")
```
